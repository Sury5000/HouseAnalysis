```python

#Selecting first features like price ,square feet kiving,bathrooms,floors sizes attributes which are must be cleaned


outlier_feat = ['price', 'sqft_living', 'sqft_lot', 'bathrooms', 'floors', 'sqft_above', 'sqft_basement']

def remove_outliers(house1, outlier_feat):
    for col in outlier_feat:
        Q1 = house1[col].quantile(0.25)
        Q3 = house1[col].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        house1 = house1[(house1[col] >= lower_bound) & (house1[col] <= upper_bound)]

    return house1

house1_cleaned = remove_outliers(house1, outlier_feat)

print(f"Original dataset size: {house1.shape[0]}")
print(f"After outlier removal: {house1_cleaned.shape[0]}")
```

```python
#Checking for distribution after outlier removal
import matplotlib.pyplot as plt

house1_cleaned[outlier_feat].hist(figsize=(8,6), bins=30, edgecolor="black")
plt.suptitle("Feature Distributions After Outlier Removal", fontsize=16)
plt.show()
```

```python
#There are some skewness on right side so we apply log transformation to minimize the variance for price and sqft_above

house1_cleaned['price'] = np.log1p(house1_cleaned['price'])
house1_cleaned['sqft_above'] = np.log1p(house1_cleaned['sqft_above'])
```

```python
house1_cleaned[outlier_feat].hist(figsize=(8,6), bins=30, edgecolor="black")
plt.suptitle("Feature Distributions After Outlier Removal", fontsize=16)
plt.show()
```

```python
#Feature Engineering by obtain the price of squareFeet by dividing the living space by price

house1_cleaned['price_per_Sqft'] = house1_cleaned['price'] / house1_cleaned['sqft_living']
house1_cleaned['price_per_Sqft']
```

```python
zip_price = house1_cleaned.groupby('zipcode')['price'].mean().sort_values(ascending = False)
zip_price
```

```python
house1['date'] = pd.to_datetime(house1['date'], format='%Y%m%dT%H%M%S')

house1['year_sold'] = house1['date'].dt.year
avg_price = house1.groupby('year_sold')['price'].mean()
```

```python
#Plotting the Average price of the house sold over years which is steadily increasing due time

plt.figure(figsize=(10, 5))
plt.plot(avg_price.index, avg_price.values, marker='o', linestyle='-')
plt.xlabel("Year Sold")
plt.ylabel("Average Price")
plt.title("Average House Price per Year Sold")
plt.grid(True)
plt.show()
```
